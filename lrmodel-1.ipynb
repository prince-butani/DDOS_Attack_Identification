```python
# Helper Functions
def check_duplicates(z):
  dups = 0
  for i in z.duplicated():
    if i == True:
      dups += 1
  print('We have {} duplicates'.format(dups))
  return dups
```

# Random UnderSampling

Since we are imbalance and resources are scarce, we shall use random undersampling for dataset to become balanced

```python
from collections import Counter
from imblearn.under_sampling import RandomUnderSampler
```

```python
X = df.drop(['Label'], axis=1)
y = df['Label']
del df
gc.collect()
```

```python
print('Original dataset shape %s' % Counter(y))
```

```python
rus = RandomUnderSampler(random_state=4869)
```

```python
X_res, y_res = rus.fit_resample(X, y)
```

```python
print('Resampled dataset shape %s' % Counter(y_res))
```

```python
del X
del y
gc.collect()
```

# Logistic Regression Model

Splitting into train & test

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=10)
```

Train the model

```python
lr = LogisticRegression(max_iter=10000)
lr.fit(X_train, y_train)
```

Predict with Test Dataset

```python
y_pred = lr.predict(X_test)
```

# Model Evaluation

Confusion Matrix

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
cnf_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sn.heatmap(cnf_matrix, annot=True)
```

# Accuracy Score

```python
accuScore = metrics.accuracy_score(y_test, y_pred)
print('Accuracy Score : ', accuScore)
print('Accuracy in percentage : ', int(accuScore*100), '%')
```
